```python
import pandas as pd
import numpy as np

# --- 1) Normalizar fechas en retornos ---
ret = retornos.copy()
ret["Dates"] = pd.to_datetime(ret["Dates"], errors="coerce").dt.normalize()

# --- 2) Normalizar fechas en barings y quedarnos con Total_return ---
bar = barings.copy()
# En tu tabla se ve "Fecha" y "Total_return"
bar["Fecha"] = pd.to_datetime(bar["Fecha"], errors="coerce").dt.normalize()

# Asegura que sea numérico
bar["Total_return"] = pd.to_numeric(bar["Total_return"], errors="coerce")

# Nos quedamos con lo mínimo y quitamos duplicados por fecha (por si existieran)
bar_map = (
    bar[["Fecha", "Total_return"]]
    .dropna(subset=["Fecha"])
    .drop_duplicates(subset=["Fecha"], keep="last")
    .set_index("Fecha")["Total_return"]
)

# --- 3) Mapear por fecha y rellenar Private Debt ---
# (esto no depende del orden de las filas)
mapped = ret["Dates"].map(bar_map)

# Rellena solo donde Private Debt esté NaN
ret["Private Debt"] = ret["Private Debt"].fillna(mapped)

# Si quieres SOBREESCRIBIR toda la columna con lo de barings (en vez de fillna), usa:
# ret["Private Debt"] = mapped

# --- 4) Dejar el dataframe en la misma forma y orden original ---
retornos = ret

# (opcional) Chequeo rápido
print("NaNs en Private Debt después:", retornos["Private Debt"].isna().sum())
print("Fechas matcheadas:", mapped.notna().sum())
retornos = retornos.bfill()
retornos.to_excel('Private_debt_filled.xlsx', index=False)
```
